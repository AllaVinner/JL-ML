```python
import pandas as pd
from os.path import join
import seaborn as sns
sns.set_palette('colorblind')
```

```python
model_names = ['vae_rf_1',
               'vae_rf_10',
               'vae_rf_100',
               'vae_rf_1000',
               'vae_rf_10000',
                ]
df_frames = [pd.read_csv(join(saved_path, model_name,'latent_df.csv')) for model_name in model_names]
latent_df = pd.concat(df_frames, ignore_index=True)
df_frames = [pd.read_csv(join(saved_path, model_name,'dimension_df.csv')) for model_name in model_names]
dimension_df = pd.concat(df_frames, ignore_index=True)
```
